## FNIR

```python
# Read the CSV files
fake_news = pd.read_csv("../datasets/Health/COVID19-FNIR/fakeNews.csv")
true_news = pd.read_csv("../datasets/Health/COVID19-FNIR/trueNews.csv")
```

```python
fake_news.head()
```

```python
#fake news drop Date Posted, Link, Region, Country, Explanation, Origin, Origin_URL, Fact_checked_by, Poynter_Label
fake_news = fake_news.drop(columns=['Date Posted', 'Link', 'Region', 'Country', 'Explanation', 'Origin', 'Origin_URL', 'Fact_checked_by', 'Poynter_Label'])
fake_news.head()
```

```python
#rename the columns text to claim and Binary Label to label
fake_news = fake_news.rename(columns={'Text': 'claim', 'Binary Label': 'label'})
fake_news.head()
```

```python
true_news.head()
```

```python
#true news drop Date Posted	Link	Region	Username	Publisher
true_news = true_news.drop(columns=['Date Posted', 'Link', 'Region', 'Username', 'Publisher'])
true_news.head()
```

```python
#true news rename the columns Text to claim and Label to label
true_news = true_news.rename(columns={'Text': 'claim', 'Label': 'label'})
true_news.head()
```

```python
# Combine both of the files
fnir = pd.concat([fake_news, true_news])
```

```python
#merge fake_new_df and fnir
health_dataset = pd.concat([fake_new_df, fnir])
```

```python
#remove duplicates
health_dataset = health_dataset.drop_duplicates(subset='claim')

#remove rows with missing values
health_dataset = health_dataset.dropna()
```

```python
#display details of merged_df such as shape, info, describe, head unique values in label column
print(health_dataset.shape)
print(health_dataset.info())
print(health_dataset.describe())
print(health_dataset.head())
print(health_dataset['label'].unique())

#display average claim length
health_dataset['claim_length'] = health_dataset['claim'].apply(lambda x: len(x.split()))
print(health_dataset['claim_length'].mean())
```

```python
#Remove punctuations
#Convert text to tokens
#Remove tokens of length less than or equal to 3
#Remove stopwords using NLTK corpus stopwords list to match
#Apply lemmatization

import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if len(word) > 3]
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

health_dataset['claim'] = health_dataset['claim'].apply(clean_text)
health_dataset.head()
```

```python
#display average claim length
health_dataset['claim_length'] = health_dataset['claim'].apply(lambda x: len(x.split()))
print(health_dataset['claim_length'].mean())
#remove claim length column
health_dataset = health_dataset.drop(columns=['claim_length'])
```

```python
#save to preprocessed_datasets folder
health_dataset.to_csv('../preprocessed_datasets/health_dataset.csv', index=False)
```
